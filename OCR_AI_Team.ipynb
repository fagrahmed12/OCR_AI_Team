```python
import cv2
from ArabicOcr import arabicocr
image_path='arabic.jpg'           # The image we need to apply the ocr to
out_image='out_arabic.jpg'        # the output of image after applyed ocr
results=arabicocr.arabic_ocr(image_path,out_image)
print(results)
words=[]
for i in range(len(results)):	
		word=results[i][1]
		words.append(word)
with open ('file_arabic.txt','w',encoding='utf-8')as myfile: #to print the out put of image as text file
		myfile.write(str(words))
```
